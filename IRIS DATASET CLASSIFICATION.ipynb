# IRIS DATASET CLASSIFICATION

Here Logistic Regression model as well as support vecctor classification model is used to classify the flowers into three categories i.e, setosa, versicolor, virginica using the sepal width, sepal length, petal width and petal length.

# Split the dataset into features and target

The dataset is split into two i.e, features and target. Here sepal length and width , petal length and width will be the features.

```python
X = df.drop('class', axis=1)
Y = df['class']
```

```python
X
```

```python
Y
```

# Train - Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)
```

# Choosing models

Here Logistic Regression as well as Support Vector Classification models are used. Since classification is the main task here Classification algorithms are the best.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

# Model training and prediction

```python
# Logistic Regression model


lr_model = LogisticRegression(max_iter= 1000)
lr_model.fit(X_train,Y_train)
lr_pred = lr_model.predict(X_test)
```

```python
# SVC model to implement SVM

svm_model = SVC()
svm_model.fit(X_train, Y_train)
svm_pred = svm_model.predict(X_test)
```

```python
# Logistic Regression models prediction

lr_df = pd.DataFrame({ 'Actual': Y_test, 'Predicted': lr_pred})
lr_df
```

```python
# SVC models prediction

svm_df = pd.DataFrame({ 'Actual': Y_test, 'Predicted': svm_pred})
svm_df
```

# Accuracy of the models

```python
from sklearn.metrics import accuracy_score, classification_report

lr_accuracy = accuracy_score(Y_test, lr_pred)
lr_report= classification_report(Y_test, lr_pred, target_names=['Iris-setosa','Iris-versicolor','Iris-virginica'])
```

```python
lr_accuracy   #Logistic Regression models Accuracy
```

```python
lr_report
```

```python
svm_accuracy = accuracy_score(Y_test, svm_pred)
svm_report= classification_report(Y_test, svm_pred)
```

```python
svm_accuracy        #SVM models Accuracy
```

```python
svm_report
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, y_pred, class_names, title):
    cm = confusion_matrix(y_true, y_pred)
    cm_df = pd.DataFrame(cm, index=class_names, columns=class_names)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_df, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Plot confusion matrix for Logistic Regression
plot_confusion_matrix(Y_test, lr_pred,['Iris-setosa','Iris-versicolor','Iris-virginica'] , title='Logistic Regression Confusion Matrix')

# Plot confusion matrix for SVM
plot_confusion_matrix(Y_test, svm_pred,['Iris-setosa','Iris-versicolor','Iris-virginica'] , title='Support Vector Machine Confusion Matrix')
```
